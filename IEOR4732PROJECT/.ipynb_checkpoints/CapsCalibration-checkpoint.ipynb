```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
from scipy.stats import norm
from matplotlib import cm
import math
```

```python
## Prepared for real data
def marketCaplet(notion,spot,maturity,tenor,disBond,forward,strike,vol):
    d1 = (1.0/(vol*math.sqrt(maturity-spot)))*math.log(forward/strike)+0.5*vol*math.sqrt(maturity-spot)
    d0 = d1 - vol*math.sqrt(maturity-spot)
    return notion*disBond*tenor*(forward*norm.cdf(d1)-strike*norm.cdf(d0))

def marketFloorlet(notion,spot,maturity,tenor,disBond,forward,strike,vol):
    d1 = (1.0/(vol*math.sqrt(maturity-spot)))*math.log(forward/strike)+0.5*vol*math.sqrt(maturity-spot)
    d0 = d1 - vol*math.sqrt(maturity-spot)
    return notion*disBond*tenor*(-forward*norm.cdf(-d1)+strike*norm.cdf(-d0))

def bondPrice(ytm,maturity):
    return 1/(1+ytm*maturity/100)
```

### solve for implied vol w.r.t strike

```python
Notion = 25000000
ytmT = [1/12,2/12,3/12,6/12,1,2,3]
ytmR = [1.56,1.56,1.59,1.59,1.58,1.66,1.69]

M = [2,3,2,3,2,3,2,3]
K = [2,2,2.5,2.5,3,3,3.5,3.5]
marketCap = [37000,83000,17000,39000,13000,21000,11000,16000]

def solversig1(sig):
    sig1=sig[0]
    sig2=sig[1]
    sig3=sig[2]
    sig4=sig[3]
    P1 = sum(marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.02,sig1) for i in range(1,25))
    P2 = sum(marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.025,sig2) for i in range(1,25))
    P3 = sum(marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.03,sig3) for i in range(1,25))
    P4 = sum(marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.035,sig4) for i in range(1,25))
    return [(P1-37000),(P2-17000),(P3-13000),(P4-11000)]

sigg = fsolve(solversig1,np.array([0.3,0.3,0.3,0.3]))
print(sigg)


     
```

### form caplets test data

```python
CapLetsMarket1 = [marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.02,sigg[0]) for i in range(1,25)]
CapLetsMarket2 = [marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.025,sigg[1]) for i in range(1,25)]
CapLetsMarket3 = [marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.03,sigg[2]) for i in range(1,25)]
CapLetsMarket4 = [marketCaplet(Notion,0,i/12,1/12,bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12),fList1[i-1],0.035,sigg[3]) for i in range(1,25)]

CapLetsMarket = CapLetsMarket1 + CapLetsMarket2 + CapLetsMarket3 + CapLetsMarket4
BtTL = 4*[bondPrice(np.interp(i/12,ytmT,ytmR),i/12) for i in range(1,25)]
BtsL = 4*[bondPrice(np.interp((i+1)/12,ytmT,ytmR),(i+1)/12) for i in range(1,25)]
TL = 4*[i/12 for i in range(1,25)]
KL = 24*[0.02]+24*[0.025]+24*[0.03]+24*[0.035]
```

### object function

```python

def objectFuncCaplet(params, Notion,Tenor,tList, kList, marketList, BtTList, BtsList,model):
    L = len(tList)
    if model == "LFM":
        N = len(params)
        prices=[pg.capletLFM(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],params[i%N]) for i in range(L)]
        return sum(kList[i]**3*(prices[i]-marketList[i])**2 for i in range(L))
    kappa = params[0]
    theta = params[1]
    sigma = params[2]
    weight = 0
    if model=="CIR":
        weight = 2.5
        prices = [pg.capletCIR(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],kappa,theta,sigma) for i in range(L)]
    elif model == "Vasicek":
        weight = 2.5
        prices = [pg.capletVasicek(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],kappa,theta,sigma) for i in range(L)]
    elif model == "Longstaff":
        prices = [pg.capletLongstaff(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],kappa,theta,sigma) for i in range(L)]
        
    return sum(kList[i]**weight*(prices[i]-marketList[i])**2 for i in range(L))

def objectFuncFloorlet(params, Notion,Tenor,tList, kList, marketList, BtTList, BtsList,model):
    L = len(tList)
    if model == "LFM":
        N = len(params)
        prices=[pg.floorletLFM(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],params[i%N]) for i in range(L)]
        return sum((prices[i]-marketList[i])**2 for i in range(L))
    kappa = params[0]
    theta = params[1]
    sigma = params[2]
    
    if model=="CIR":
        prices = [pg.floorletCIR(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],kappa,theta,sigma) for i in range(L)]
    elif model == "Vasicek":
        prices = [pg.floorletVasicek(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],kappa,theta,sigma) for i in range(L)]
    elif model == "Longstaff":
        prices = [pg.floorletLongstaff(0,tList[i],tList[i]+Tenor,Notion,kList[i],BtTList[i],BtsList[i],kappa,theta,sigma) for i in range(L)]
            
    return sum((prices[i]-marketList[i])**2 for i in range(L))

    
```

### Calibration


```python
Model = "CIR"
```

```python
#Calibration
params=None
if Model =="CIR":
#     params = np.array([8,0.01,0.12])
    params = np.array([2,0.01,0.3])
elif Model =="Vasicek": 
    params = np.array([0.6,0.02,0.3])
elif Model =="LFM":
    params=np.array([0.6,0.4,0.4,0.4,0.4,0.4,0.4,0.35,0.34,0.34,0.34,0.3,0.28,0.28,0.28,0.28,0.28,0.27,0.26,0.26,0.26,0.26,0.26,0.3])
    #params=np.array([0.6-_*(0.4)/24 for _ in range(24)])
#params = np.array([0.6,0.02,0.3])
elif Model == "Longstaff":
    params = np.array([0,1,0.4])
maxN = 10000
if Model == "LFM":
    maxN = maxN*240000
opt = {'maxiter':maxN, 'maxfev':5e3}
sol = minimize(objectFuncCaplet, params, args = (Notion,1/12,TL,KL,CapLetsMarket,BtTL,BtsL,Model), method='Nelder-Mead', options=opt)

print(sol.message)
```

```python
optParam = np.array(sol.x)
print(optParam)
kappa = optParam[0] #r0 for longstaff
theta = optParam[1] #k for longstaff
sig = optParam[2]   #sigma for longstaff
```
